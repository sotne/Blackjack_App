```python
import random

class Card:
    def __init__(self, suit, value):
        self.suit = suit
        self.value = value
        
    def __repr__(self):
        return f'{self.value} of {self.suit}'
    
class Deck:
    def __init__(self):
        self.cards = [Card(suit, value) for suit in ['hearts', 'diamonds', 'clubs', 'spades'] for value in ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'jack', 'queen', 'king', 'ace']]
        
    def shuffle(self):
        random.shuffle(self.cards)
        
    def deal(self):
        return self.cards.pop()
    
class Hand:
    def __init__(self):
        self.cards = []
        
    def hit(self, card):
        self.cards.append(card)
        
    def __repr__(self):
        return f'{self.cards}'
    
class Blackjack:
    def __init__(self):
        
        self.deck = Deck()
        self.player_hand = []
        self.dealer_hand = []
        self.deck.shuffle()
        self.player_hand.hit(self.deck.deal())
        self.player_hand.hit(self.deck.deal())
        self.dealer_hand.hit(self.deck.deal())
        self.dealer_hand.hit(self.deck.deal())
    
    def get_point_value(self, card):
        if card.value == 'ace':
            return 11
        elif card.value in ['jack', 'queen', 'king']:
            return 10
        else:
            return int(card.value)    
    
    def play_round(self):        
        print(f'Your hand:{self.player_hand}')
    
        
        while True:
            choice = input('Would you like to hit or stand? ')
            if choice == 'hit':
                self.player_hand.hit(self.deck.deal())
                print(f'Your hand:{self.player_hand}')
                if self.determine_winner() == 'player':
                    print('You won!')
                    return
                elif self.determine_winner() == 'dealer':
                    print('You lost!')
                    return
            elif choice == 'stand':
                break
        
        while True:
            self.dealer_hand.hit(self.deck.deal())
            print(f'Dealer hand:{self.dealer_hand}')  
            if self.determine_winner() == 'player':
                print('You won!')
                return
            elif self.determine_winner() == 'dealer':
                print('You lost!')
                return
        
    def determine_winner(self):
        player_total = sum([self.get_point_value(card) for card in self
                                    .player_hand.cards])
        dealer_total = sum([self.get_point_value(card) for card in self
                            .dealer_hand.cards])
        
        if player_total == 21:
            return 'player'
        elif dealer_total == 21:
            return 'dealer'
        elif player_total > 21:
            return 'dealer'
        elif dealer_total > 21:
            return 'player'
        else:
            return 'none'

        
game = Blackjack()
game.play_round()



                            
```
